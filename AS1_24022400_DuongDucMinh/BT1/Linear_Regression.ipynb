# Linear Regression
Part 1: Bài 3: linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data: Height (cm) và Weight (kg)
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
y = np.array([[49, 50, 51, 54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T

# Chuyển về vector 1 chiều
x = X.flatten()
y = y.flatten()

# Số điểm dữ liệu
N = x.shape[0]

# Tính hệ số m (slope) và b (intercept)
m = (N * np.sum(x * y) - np.sum(x) * np.sum(y)) / (N * np.sum(x ** 2) - (np.sum(x) ** 2))
b = (np.sum(y) - m * np.sum(x)) / N

print("Slope (m):", m)
print("Intercept (b):", b)

# Tạo đường hồi quy
x_min = np.min(x)
y_min = m * x_min + b

x_max = np.max(x)
y_max = m * x_max + b

# Vẽ hình
fig, ax = plt.subplots()

sns.scatterplot(
    x=x,
    y=y,
    ax=ax,
    alpha=0.7
)

sns.lineplot(
    x=[x_min, x_max],
    y=[y_min, y_max],
    linewidth=2,
    color='red'
)

plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Linear Regression: Height vs Weight")

plt.show()
```
