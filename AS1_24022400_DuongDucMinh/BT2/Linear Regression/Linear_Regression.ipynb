# Linear Regression
P4. Logistic regression (trang 66)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Đọc dữ liệu

```python
df = pd.read_csv('data_linear.csv')
df.columns = ['mét vuông', 'giá']
df.head()
```

## Vẽ scatter plot để xem dữ liệu

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x='mét vuông',
    y='giá'
)

plt.show()
```

## Lấy dữ liệu x và y

```python
x = df['mét vuông'].values
y = df['giá'].values
```

## Tính hệ số hồi quy m và b

```python
N = x.shape[0]

m = (N * np.sum(x * y) - np.sum(x) * np.sum(y)) / \
    (N * np.sum(x ** 2) - (np.sum(x) ** 2))

b = (np.sum(y) - m * np.sum(x)) / N

print('M:', m)
print('B:', b)
```

## Tính đường regression

```python
x_min = np.min(x)
y_min = m * x_min + b

x_max = np.max(x)
y_max = m * x_max + b
```

## Vẽ scatter plot và regression line

```python
fig, ax = plt.subplots()

sns.scatterplot(
    data=df,
    x='mét vuông',
    y='giá',
    ax=ax,
    alpha=0.4
)

sns.lineplot(
    x=[x_min, x_max],
    y=[y_min, y_max],
    linewidth=1.5,
)

plt.show()
```

## Dự đoán giá nhà

```python
x1 = 50
y1 = m * x1 + b

print(f'Gia nha cho {x1}m^2 la: {y1}')
```
